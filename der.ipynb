```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

im = np.array(Image.open('empire.jpg'))
plt.imshow(im)
x = [100,100,400,400]
y = [200,500,200,500]
plt.plot(x, y, 'ks:') # şekil ve renk değişir 'go r* -go ks:'
#plt.plot(x[:2], y[:2])
plt.title('Plotting: "empire.jpg"')
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure()
plt.hist(im.flatten(),128) # 2-3 boyutlu resmi tek boyuta indirir
plt.title('Histogram: "HAHO.jpg"')
plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

im = np.array(Image.open('empire.jpg').convert('L')) # gri tonlama için convert('L') ekledim
plt.figure()
plt.contour(im, origin='image')
plt.title('Contours: "HAHO.JPG"')
plt.show()
plt.axis('equal') # Eksen oranlarını eşit yapar
plt.axis('off')   # Eksenleri kapatır x-y ölçüleri gider



```
